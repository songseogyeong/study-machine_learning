#### ğŸ“Š LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

# ë¬¸ìì—´ í”¼ì²˜ ì¶”ì¶œ
columns = ['ì„±ë³„', 'ì¢…ë¥˜']
# ê° ì»¬ëŸ¼ì˜ LabelEncoder ê°ì²´ë¥¼ ì €ì¥í•  ë”•ì…”ë„ˆë¦¬(encoders) ì´ˆê¸°í™”
encoders = {}

# ì»¬ëŸ¼ ë°˜ë³µí•˜ì—¬ columnì— ë‹´ì•„ ì ìš©:
for column in columns:
    # LabelEncoder ê°ì²´ ìƒì„±, ë¬¸ìì—´ ë°ì´í„° ì¸ì½”ë”©ì„ í†µí•´ ì •ìˆ˜ë¡œ í˜•ë³€í™˜ í›„ ë°ì´í„° êµì²´
    encoder = LabelEncoder()
    # LabelEncoderë¥¼ ì‚¬ìš©í•˜ì—¬ ê° ì»¬ëŸ¼ì˜ ê°’ì„ ë³€í™˜
    result = encoder.fit_transform(pre_c_df[column])
    # ë³€í™˜ëœ ê°’ ì ìš©
    pre_c_df[column] = result
    # ë³€í™˜ëœ ê°’ì˜ í´ë˜ìŠ¤(ê³ ìœ í•œ ê°’) ì €ì¥
    encoders[column] = encoder.classes_

# ê³ ìœ í•œ ê°’ í™•ì¸
print(encoders)
```

#### ğŸ“Š VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
# VIF í™•ì¸
get_vif(features)
```

#### ğŸ“Š íšŒê·€ ë¶„ì„

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
# í”¼ì²˜, íƒ€ê²Ÿ ë°ì´í„° ë¶„ë¦¬
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# ë¬¸ì œ/ì •ë‹µ ë° í•™ìŠµ/í›ˆë ¨ ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ì„ í˜• íšŒê·€ ëª¨ë¸
l_r = LinearRegression()
# í›ˆë ¨: ëª¨ë¸ì´ ìµœì ì˜ ê°€ì¤‘ì¹˜ì™€ í¸í–¥ì„ í•™ìŠµ
l_r.fit(X_train.values, y_train.values)
```

```python
# ì˜ˆì¸¡
prediction = l_r.predict(X_test.values)
# í‰ê°€
get_evaluation(y_test.values, prediction)
```

ë¹„ì„ í˜• ë°ì´í„°ë¥¼ í•™ìŠµí•˜ê¸° ìœ„í•´ ë‹¤ì°¨ì› ì‹ì„ ë§Œë“œëŠ” ê¸°ë²•

```python
from sklearn.preprocessing import PolynomialFeatures

# ì°¨ì› í™•ì¥, ë¹„ì„ í˜• ê´€ê³„ë¥¼ ëª¨ë¸ë§
# degree: ì°¨ìˆ˜ ì¡°ì ˆ (2ë¶€í„° ìƒìŠ¹ì‹œí‚¤ê¸°, ê¸°ì¡´ì˜ í”¼ì²˜ë¥¼ ìœ ì§€í•˜ë©° ì„¤ì •í•œ ì°¨ìˆ˜ë¡œ ë§Œë“¦)
poly_features = PolynomialFeatures(degree=3).fit_transform(features)

# ë¬¸ì œ/ì •ë‹µ ë° í•™ìŠµ/í›ˆë ¨ ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(poly_features, targets, test_size=0.2, random_state=124)

# ì„ í˜• íšŒê·€ ëª¨ë¸
l_r = LinearRegression()
# í›ˆë ¨: ëª¨ë¸ì´ ìµœì ì˜ ê°€ì¤‘ì¹˜ì™€ í¸í–¥ì„ í•™ìŠµ
l_r.fit(X_train, y_train)
```

```python
# ì˜ˆì¸¡
prediction = l_r.predict(X_test)
# í‰ê°€
get_evaluation(y_test.values, prediction)
```

#### ğŸ“Š decisiontree regression

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

# ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
# í”¼ì²˜, íƒ€ê²Ÿ ë°ì´í„° ë¶„ë¦¬
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# ë¬¸ì œ/ì •ë‹µ ë° í•™ìŠµ/í›ˆë ¨ ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# íšŒê·€ ëª¨ë¸
dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor()
lgb_r = LGBMRegressor(n_estimators=100)

# ê° ëª¨ë¸ ë‹´ê¸°
models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]

# ë°˜ë³µí•˜ì—¬ ëª¨ë¸ì— ëŒ€í•œ ë‹¤ìŒê³¼ ê°™ì€ ì‘ì—… ìˆ˜í–‰:
for model in models:
    # ëª¨ë¸ í•™ìŠµ
    model.fit(X_train, y_train)
    # ëª¨ë¸ ì˜ˆì¸¡
    prediction = model.predict(X_test)
    # ëª¨ë¸ ì´ë¦„
    print(model.__class__.__name__)
    # í‰ê°€
    get_evaluation(y_test, prediction)
```

#### ğŸ“Š GridSearchCV
- ê°€ì¥ ìš°ìˆ˜í•œ ëª¨ë¸ì˜ í•˜ì´í¼ íŒŒë¼ë¯¸í„° ì¡°ì •

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

# ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
# í”¼ì²˜, íƒ€ê²Ÿ ë°ì´í„° ë¶„ë¦¬
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# ë¬¸ì œ/ì •ë‹µ ë° í•™ìŠµ/í›ˆë ¨ ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ê°€ì¥ ì„±ëŠ¥ì´ ìš°ìˆ˜í•œ ëª¨ë¸ ë‹´ê¸°
rf_r = RandomForestRegressor(random_state=124)

# íŒŒë¼ë¯¸í„° ê°’ ì¡°ì •
parameters = {'max_depth': [4, 8, 12, 20], 'min_samples_split': [20, 30, 40, 50, 60], 'n_estimators': [10, 50, 100, 500, 1000]}
# êµì°¨ê²€ì¦
# n_splits: ë°ì´í„°ë¥¼ ëª‡ ê°œì˜ í´ë“œë¡œ ë‚˜ëˆŒì§€ë¥¼ ê²°ì • (ì¼ë°˜ì ìœ¼ë¡œ 5 ë˜ëŠ” 10)
# shuffle: ë¶„í•  ì „ ë°ì´í„° í˜¼í•© ì—¬ë¶€ 
kfold = KFold(n_splits=10, random_state=124, shuffle=True)

# í•™ìŠµ ë° êµì°¨ ê²€ì¦ ëª¨ë¸ ì„¤ì •
grid_rf_r = GridSearchCV(rf_r, param_grid=parameters, scoring='neg_mean_squared_error', cv=kfold, n_jobs=-1)
# grid_rf_r = GridSearchCV(rf_r, param_grid=parameters, scoring='r2', cv=kfold)

# í›ˆë ¨
grid_rf_r.fit(X_train, y_train)
```

```python
# í›ˆë ¨ ê²°ê³¼ í™•ì¸
result_df = pd.DataFrame(grid_rf_r.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```
