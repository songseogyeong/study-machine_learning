- ì„œí¬íŠ¸ ë²¡í„°ë“¤ì´ ê²°ì • ê²½ê³„ (Decision boundary)ë¥¼ ê²°ì •í•œë‹¤.

- ìƒˆë¡œìš´ ë°ì´í„°ê°€ ë“¤ì–´ì˜¤ë”ë¼ë„ ê²°ì • ê²½ê³„ë¥¼ ì¤‘ì‹¬ìœ¼ë¡œ ë‘ ì§‘ë‹¨ì´ ë©€ë¦¬ ë–¨ì–´ì ¸ ìˆì–´ì•¼ ë‘ ì§‘ë‹¨ì„ ì˜ êµ¬ë¶„í•  ìˆ˜ ìˆê¸° ë•Œë¬¸ì— ì¼ë°˜í™”í•˜ê¸° ì‰¬ì›Œì§„ë‹¤.
- ì˜ˆì¸¡ ë³€ìˆ˜ì˜ ì°¨ì›ë³´ë‹¤ í•œ ì°¨ì› ë‚®ì•„ì§€ë©°, Nì°¨ì› ê³µê°„ì—ì„œ í•œ ì°¨ì› ë‚®ì€ N-1ì°¨ì›ì˜ ê²°ì • ê²½ê³„ê°€ ìƒê¸´ë‹¤.  
  ì¦‰, 2ì°¨ì› ê³µê°„ì—ì„œ ê²°ì • ê²½ê³„ëŠ” ì„ ìœ¼ë¡œ ê²°ì •ë˜ê³ , ê³ ì°¨ì›ì—ì„œëŠ” ê²°ì • ê²½ê³„ëŠ” ì„ ì´ ì•„ë‹Œ í‰ë©´ ì´ìƒì˜ ë„í˜•ì´ë©°, ì´ë¥¼ "ì´ˆí‰ë©´(Hyperplane)"ì´ë¼ê³  ë¶€ë¥¸ë‹¤.

#### í•˜ë“œ ë§ˆì§„(Hard margin)
- ë§¤ìš° ì—„ê²©í•˜ê²Œ ì§‘ë‹¨ì„ êµ¬ë¶„í•˜ëŠ” ë°©ë²•ìœ¼ë¡œ ì´ìƒì¹˜ë¥¼ í—ˆìš©í•´ ì£¼ì§€ ì•ŠëŠ” ë°©ë²•ì´ë‹¤.
- ì´ìƒì¹˜ë¥¼ í—ˆìš©í•˜ì§€ ì•Šê¸° ë•Œë¬¸ì— ê³¼ì í•©ì´ ë°œìƒí•˜ê¸° ì‰½ê³ , ìµœì ì˜ ê²°ì • ê²½ê³„ë¥¼ ì˜ ëª» êµ¬ë¶„í•˜ê±°ë‚˜ ëª» ì°¾ëŠ” ê²½ìš°ê°€ ìƒê¸¸ ìˆ˜ ìˆë‹¤.
- C(cost)ëŠ” íŒ¨ë„í‹°ë¥¼ ì¡°ì ˆí•  ìˆ˜ ìˆê³ , ê°’ì´ ì»¤ì§ˆìˆ˜ë¡ ê²°ì • ê²½ê³„ê°€ ë°ì´í„°ì— ë” ì •í™•í•˜ê²Œ ë§ì¶°ì§„ë‹¤.
- Cë¥¼ ë‚®ì¶”ë©´ ì¼ì„ ëœí•˜ê²Œ í•˜ëŠ” ê²ƒì´ê³ , Cë¥¼ ë†’ì´ë©´ ì¼ì„ ë”í•´ì„œ ë” ì„¬ì„¸í•˜ê²Œ ì°¾ì•„ë‚¸ë‹¤.
- Cê°€ ë„ˆë¬´ ë‚®ìœ¼ë©´ underfitting ë  ê°€ëŠ¥ì„±ì´ ì»¤ì§€ê³ , Cê°€ ë„ˆë¬´ ë†’ìœ¼ë©´ overfittingì´ ë°œìƒí•  ìˆ˜ ìˆë‹¤.

#### ì†Œí”„íŠ¸ ë§ˆì§„(Soft margin)
- ì´ìƒì¹˜ë¥¼ í—ˆìš©í•´ì„œ ì¼ë¶€ ë°ì´í„°ë¥¼ ì˜ëª» ë¶„ë¥˜í•˜ë”ë¼ë„ ë‚˜ë¨¸ì§€ ë°ì´í„°ë¥¼ ë”ìš± ì˜ ë¶„ë¥˜í•´ ì£¼ëŠ” ë°©ë²•ì´ë‹¤.
- ì´ìƒì¹˜ í—ˆìš©ìœ¼ë¡œ ì¸í•´ ë°ì´í„°ì˜ íŒ¨í„´ì„ ì˜ ê°ì§€í•˜ì§€ ëª»í•˜ëŠ” ë¬¸ì œì ì´ ìƒê¸¸ ìˆ˜ ìˆë‹¤.

> Cê°€ í´ ìˆ˜ë¡ loss functionì—ì„œ ì˜¤ì°¨í•­ì¸ Î¾<sub>i</sub>ì˜ ì˜í–¥ë ¥ì´ ì»¤ì§€ê²Œ ë˜ê¸° ë•Œë¬¸ì— ë§ˆì§„ì˜ í¬ê¸°ê°€ ì¤„ì–´ë“¤ê³ (í•˜ë“œ ë§ˆì§„),  
> ë°˜ëŒ€ë¡œ Cê°€ ì‘ì„ ìˆ˜ë¡ ë§ˆì§„ì˜ í¬ê¸°ê°€ ëŠ˜ì–´ë‚œë‹¤(ì†Œí”„íŠ¸ ë§ˆì§„). ì ì ˆíˆ ì¡°ì ˆí•˜ë©´ ì˜¤íˆë ¤ ì„±ëŠ¥ì´ ì¢‹ì•„ì§ˆ ìˆ˜ ìˆë‹¤.

#### ì»¤ë„ íŠ¸ë¦­ (Kernel trick)
- ì„ í˜•ìœ¼ë¡œ ì™„ì „íˆ ë¶„ë¥˜í•  ìˆ˜ ì—†ëŠ” ë°ì´í„° ë¶„í¬ê°€ ìˆì„ ê²½ìš° ì†Œí”„íŠ¸ ë§ˆì§„ì„ í†µí•´ ì–´ëŠì •ë„ ì˜¤ë¥˜ëŠ” í—ˆìš©í•˜ëŠ” í˜•íƒœë¡œ ë¶„ë¥˜í•  ìˆ˜ëŠ” ìˆë‹¤.  
  í•˜ì§€ë§Œ, ë” ì˜ ë¶„ë¥˜í•˜ê¸° ìœ„í•´ì„œëŠ” ì°¨ì›ì„ ë†’ì—¬ì•¼ í•œë‹¤. ì´ë¥¼ ê³ ì°¨ì› ë§¤í•‘ì´ë¼ê³  í•˜ê³  ì´ ë•Œ ì»¤ë„ íŠ¸ë¦­ì„ ì‚¬ìš©í•œë‹¤.
- ì €ì°¨ì›ìœ¼ë¡œ í•´ê²°í•˜ê¸° ì–´ë ¤ìš´ ë¬¸ì œë“¤ì€ ê³ ì°¨ì›ìœ¼ë¡œ ë³€í™˜ì‹œì¼œ ë¬¸ì œë¥¼ í•´ê²°í•  ë•Œ ì‚¬ìš©í•œë‹¤.

- ë¹„ì„ í˜• ë°ì´í„°ì¼ ë•Œ RBF ì»¤ë„ì„ ì‚¬ìš©í•˜ê³ , ì„ í˜• ë°ì´í„°ì¼ ë•Œ linear ì»¤ë„ì„ ì‚¬ìš©í•˜ëŠ” ê²ƒì´ íš¨ê³¼ì ì´ë‹¤.
- RBF ì»¤ë„ì„ ì‚¬ìš©í•˜ê²Œ ë˜ë©´, gamma í•˜ì´í¼ íŒŒë¼ë¯¸í„°ë¥¼ ì¡°ì •í•  ìˆ˜ ìˆìœ¼ë©°, ì´ëŠ” ë°ì´í„° í¬ì¸í„°ë“¤ì˜ ì˜í–¥ ë²”ìœ„ë¥¼ ê²°ì •í•œë‹¤.
- gammaê°€ í´ìˆ˜ë¡ í•˜ë‚˜ì˜ ë²¡í„° ì˜í–¥ë ¥ ê±°ë¦¬ê°€ ì§§ì•„ì§€ê³ , gammeê°€ ì‘ì„ ìˆ˜ë¡ ê±°ë¦¬ê°€ ê¸¸ì–´ì§„ë‹¤.
- ì¦‰, gammaê°€ í´ìˆ˜ë¡ í‘œì¤€í¸ì°¨ê°€ ë‚®ì•„ì§„ë‹¤. gammaê°€ ë„ˆë¬´ ì‘ìœ¼ë©´ ê³¼ì†Œì í•©ë  ê°€ëŠ¥ì„±ì´ í¬ê³ , ë„ˆë¬´ ë†’ìœ¼ë©´ ê³¼ëŒ€ì í•©ì˜ ìœ„í—˜ì´ ìˆë‹¤.

### â…¡. ë°ì´í„° ì „ì²˜ë¦¬

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

#### ğŸ“Š íƒ€ê²Ÿ ë¹„ì¤‘ í™•ì¸

```python
# íƒ€ê²Ÿì˜ ë¹„ì¤‘ í™•ì¸
m_df.price_range.value_counts()
```

### â…¢. ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬

#### ğŸ“Š ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬

```python
# í”¼ì²˜, íƒ€ê²Ÿ ë°ì´í„° ë¶„ë¦¬
features, targets = m_df.iloc[:, :-1], m_df.iloc[:, -1]

# ë¬¸ì œ/ì •ë‹µ, í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶„
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=124)
```

### â…£. ë°ì´í„° í›ˆë ¨

#### ğŸ“Š ë¶„ë¥˜ ëª¨ë¸ ì„¤ì •

```python
# íŒŒë¼ë¯¸í„° ê°’ ì§€ì •

# CëŠ” 10ì˜ ì°¨ìˆ˜ ë§Œí¼ ê³±í•˜ê¸°
# ì˜¤ë¥¸ìª½ìœ¼ë¡œ ê°ˆìˆ˜ë¡ í•˜ë“œë§ˆì§„, ì™¼ìª½ìœ¼ë¡œ ê°ˆìˆ˜ë¡ ì†Œí”„íŠ¸ë§ˆì§„
# rbfì´ ë†’ìœ¼ë©´ ë¹„ì„ í˜• ë°ì´í„°

# íŒŒì´í”„ ë¼ì¸ìœ¼ë¡œ êµ¬ì¶•í•œ ê° íŒŒì´í”„ì˜ ì´ë¦„ì„ ë¨¼ì € ì‘ì„±í•˜ê³ ,
# (_)ì–¸ë”ë°”ë¥¼ ë‘ë²ˆ(_) ì‘ì„±í•œ ë’¤ íŒŒë¼ë¯¸í„° ì´ë¦„ì„ ì‘ì„±í•œë‹¤.
parameters = {
    'svc__C' : [0.001, 0.01, 0.1, 1, 10, 100],
    'svc__gamma' : [0.001, 0.01, 0.1, 1, 10, 100],
    'svc__kernel' : ['linear', 'rbf']
}

# ë¶„ë¥˜ ëª¨ë¸ ë‹´ê¸°
# probability: ì˜ˆì¸¡ ì‹œ í´ë˜ìŠ¤ ë³„ í™•ë¥ ì„ ì œê³µ ì—¬ë¶€
svc = SVC(probability=True)
```

```python
# 1. íŒŒì´í”„ë¼ì¸ êµ¬ì¶•
# ì „ì²˜ë¦¬ â†’ ë¶„ë¥˜
svc_pipe = Pipeline([('standard_scaler', StandardScaler()), ('svc', svc)])
```

```python
# 2. GridSearchCV
# í•™ìŠµ ë° êµì°¨ ê²€ì¦ ì„¤ì •
g_svc = GridSearchCV(svc_pipe, param_grid=parameters, cv=3, refit=True, return_train_score=True)
```

#### ğŸ“Š ë°ì´í„° í›ˆë ¨

```python
# 3. í›ˆë ¨
g_svc.fit(X_train, y_train)
```

```python
# í™•ì¸í•˜ê³  ì‹¶ì€ ì»¬ëŸ¼ë§Œ ì¶”ì¶œí•˜ì—¬ ìƒˆë¡œìš´ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ë§Œë“¤ê¸°
result_df = pd.DataFrame(g_svc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)

# Tip
# params: ëª¨ë¸ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¡°í•©
# C: ì˜¤ì°¨ í—ˆìš© ë²”ìœ„ë¥¼ ì¡°ì ˆ, gamma: ì»¤ë„ í•¨ìˆ˜ì˜ ì˜í–¥ì„ ì œì–´, kernel: ë°ì´í„°ë¥¼ ë³€í™˜í•˜ì—¬ ê³ ì°¨ì› ê³µê°„ìœ¼ë¡œ ë§¤í•‘
# mean_test_score: êµì°¨ ê²€ì¦ì„ í†µí•´ ì–»ì€ í‰ê·  í…ŒìŠ¤íŠ¸ ì ìˆ˜ (ë†’ì€ ì ìˆ˜ì¼ ìˆ˜ë¡ ì¢‹ìŒ)
# rank_test_score: í‰ê·  í…ŒìŠ¤íŠ¸ ì ìˆ˜ì— ëŒ€í•œ ìˆœìœ„
```

```python
# íŠ¹ì • ë°ì´í„° í™•ì¸
print(result_df['params'][67])
```

```python
# íŠ¹ì • ë°ì´í„°ì˜ params í™•ì¸
result_df[result_df['rank_test_score'] == 20]['params']
```

#### ğŸ“Š ë°ì´í„° ì˜ˆì¸¡

### â…¥. ì˜¤ì°¨í–‰ë ¬

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction, average='macro')
    recall = recall_score(y_test , prediction, average='macro')
    f1 = f1_score(y_test, prediction, average='macro')
    
    print('ì˜¤ì°¨ í–‰ë ¬')
    print(confusion)
    print('ì •í™•ë„: {0:.4f}, ì •ë°€ë„: {1:.4f}, ì¬í˜„ìœ¨: {2:.4f}, F1: {3:.4f}'.format(accuracy, precision, recall, f1))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
# ì˜¤ì°¨í–‰ë ¬
# í•´ë‹¹ ë°ì´í„°ëŠ” ë‹¤ì¤‘ë¶„ë¥˜ì´ë¯€ë¡œ, ROC-AUCë¥¼ ì‚¬ìš©í•˜ì§€ ì•ŠìŒ
get_evaluation(y_test, prediction, svc, X_test)
```
