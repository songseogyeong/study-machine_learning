```python
from sklearn.preprocessing import LabelEncoder

# ì¸ì½”ë” ë§¤ì†Œë“œ ë‹´ê¸°
spam_encoder = LabelEncoder()
# ë¬¸ìì—´ íƒ€ê²Ÿì„ ìˆ«ìë¡œ ì¸ì½”ë”©
targets = spam_encoder.fit_transform(spam_df.loc[:, 'Category'])
# ì¸ì½”ë”© ëœ ë°ì´í„°ë¡œ êµì²´
spam_df['Target'] = targets
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
# íŒŒì´í”„: ë°ì´í„° ì „ì²˜ë¦¬ ë° ëª¨ë¸ í•™ìŠµ ê³¼ì •ì„ ë‹¨ì¼ ê°ì²´ë¡œ ë¬¶ìŒ (ë³€í™˜ê¸°, ì¶”ì •ê¸°)
# Pipeline([('key', ê°ì²´), ('key', ê°ì²´)])

# ë§¤ë²ˆ ì‘ì—… ì‹œ Vectorizerë¥¼ ì‚¬ìš©í•´ì•¼ í•˜ì§€ë§Œ, íŒŒì´í”„ ë¼ì¸ìœ¼ë¡œ êµ¬ì¶• ì‹œ ì§ì ‘ í•  í•„ìš” ì—†ë‹¤.
# íŒŒì´í”„ 2ê°œ ì—°ê²°í•˜ì—¬ ì²˜ë¦¬ (í…ìŠ¤íŠ¸ë¥¼ ë°±í„°ë¡œ í˜•ë³€í™˜ â†’ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ë¶„ë¥˜ê¸°)
m_nb_pipe = Pipeline([('count_vectorizer', CountVectorizer()), ('multinomial_NB',  MultinomialNB())])

# Tip
# ë³€í™˜ê¸° (Transformer): ë°ì´í„° ì „ì²˜ë¦¬ ë‹¨ê³„ì—ì„œ ì‚¬ìš©
# ì¶”ì •ê¸° (Estimator): ëª¨ë¸ í•™ìŠµ ë‹¨ê³„ì—ì„œ ì‚¬ìš©
```

```python
# ë¬¸ì œ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ì˜ˆì¸¡í•˜ê¸°
prediction = m_nb_pipe.predict(X_test)
```
