#### 📊 데이터 전처리

#### ❗ 이슈
- 데이터 이상치 제거 시 데이터가 하나도 남지 않는 문제 발생으로 별도의 이상치 제거를 하지 않음.
- (도메인 지식) 유전공학 및 센서 모의 테스트 등...에서는 이상치 제거하지 않는 게 좋음.

```python
from sklearn.preprocessing import LabelEncoder

# 데이터 프레임 복제
pre_b_df = b_df.copy()

# LabelEncoder 객체 생성, 문자열 데이터 인코딩을 통해 정수로 형변환 후 데이터 교체
encoder = LabelEncoder()
# LabelEncoder를 사용하여 각 컬럼의 값을 변환
result = encoder.fit_transform(b_df['type'])
# 변환된 데이터로 대체
pre_b_df.loc[:, 'type'] = result

pre_b_df
```

```python
# 변환된 값의 클래스(고유한 값) 저장
encoder.classes_
```

```python
# 타입 컬럼 타겟이라는 명으로 생성 (위치 변경을 위함)
pre_b_df['target'] = pre_b_df.type
# 사용이 완료된 컬럼은 제거
pre_b_df = pre_b_df.drop(labels=['type'], axis=1)
```

```python
pre_b_df
```

```python
# 불필요한 피처 제거
pre_b_df = pre_b_df.drop(labels=['samples'], axis=1)
pre_b_df
```

```python
import numpy as np

# 숫자형으로 변환
pre_b_df = pre_b_df.apply(pd.to_numeric)
```

#### 📊 PCA

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# 데이터 세트 분리
# 피처, 타겟 데이터 분리
features, targets = pre_b_df.iloc[:, :-1], pre_b_df.iloc[:, -1]

# 문제/정답 및 학습/테스트 데이터 분리
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=124)

# SMOTE (over sampling)
# smote 객체 생성
smote = SMOTE(random_state=124)
# over sampling 진행
over_X_train, over_y_train = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.decomposition import PCA

# 14번 반복
for i in range(14):
    # 1차원~14차원까지 객체 생성
    pca = PCA(n_components=(i + 1))

    # 학습 데이터로 병합
    pca_train = pca.fit_transform(over_X_train)

    # 보존률 확인
    print(pca.explained_variance_ratio_.sum())
```

```python
# 차원축소 객체 생성 (2차원)
pca = PCA(n_components=2)

# 학습 데이터로 병합 
pca_train = pca.fit_transform(over_X_train)

# train
# 컬런명 설정
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]

# 차원 축소 데이터 프레임 생성
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns)
# 타겟 데이터 추가
pca_train_df.loc[:, 'target'] = over_y_train
```

```python
pca_train_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 산점도 출력
sns.scatterplot(x="pca1", y="pca2", hue='target', data=pca_train_df, alpha=0.5)
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

for i in range(4):
    lda = LinearDiscriminantAnalysis(n_components=(i + 1))
    
    lda_train = lda.fit_transform(over_X_train, over_y_train)

    print(lda.explained_variance_ratio_.sum())
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)

lda_train = lda.fit_transform(over_X_train, over_y_train)

lda_columns = [f'lda{i + 1}' for i in range(lda_train.shape[1])]
lda_train_df = pd.DataFrame(lda_train, columns=lda_columns)
lda_train_df.loc[:, 'target'] = over_y_train
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="lda1", y="lda2", hue='target', data=lda_train_df, alpha=0.5)
plt.show()
```

#### 📊 LDA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# 파라미터 값 조정
parameters = {'rfc__max_depth': [5, 10, 20], 'rfc__min_samples_split': [10, 50, 100]}

# 분류 모델 객체 생성
rfc = RandomForestClassifier()

# 파이프라인 구축
# 정규화, 차원축소, 분류 모델
pipe = Pipeline([('std', StandardScaler()),
                 ('lda', LinearDiscriminantAnalysis(n_components=2)),
                 ('rfc', rfc)])

# 하이퍼파라미터 조정
g_rfc = GridSearchCV(pipe, param_grid=parameters, cv=3, scoring='accuracy')

# 훈련
g_rfc.fit(over_X_train, over_y_train)
```

```python
# 훈련 결과 확인
result_df = pd.DataFrame(g_rfc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction, average='macro')
    recall = recall_score(y_test , prediction, average='macro')
    f1 = f1_score(y_test, prediction, average='macro')
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}'.format(accuracy, precision, recall, f1))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
# 최적의 모델 담기
rfc = g_rfc.best_estimator_

# 예측
prediction = rfc.predict(X_test)
# 평가 (오차행렬)
get_evaluation(y_test, prediction)
```
