### Dimension Reduction Task01

- 기대 수명 데이터 세트를 회귀 분석 하되, 차원을 축소하여 진행해본다.

```python
import pandas as pd

# 데이터 세트 호출
l_df = pd.read_csv('./datasets/life_expectancy.csv')
l_df
```

```python
# 불필요한 컬럼 제거
pre_l_df = l_df.drop(labels=['Country', 'Year', 'Status'], axis=1)
pre_l_df
```

```python
# 결측치 확인
pre_l_df.isna().sum()
```

```python
# 결측치가 있는 컬럼 담기
columns = pre_l_df.columns[pre_l_df.isna().any()]

# 결측치 제거
for column in columns:
    pre_l_df = pre_l_df[~pre_l_df[column].isna()]

# 결측치 제거 확인
pre_l_df.isna().sum().sum()
```

```python
# 상관관계 확인
pre_l_df.corr()['Life expectancy '].sort_values(ascending=False)[1:]
```

```python
pre_l_df
```

```python
from sklearn.preprocessing import StandardScaler

# 표준화 객체 생성
std = StandardScaler()
# 피처만 담아 표준화 진행
result = std.fit_transform(pre_l_df.iloc[:, 1:])
# 표준화 데이터 프레임 생성
std_pre_l_df = pd.DataFrame(result, columns=pre_l_df.iloc[:, 1:].columns)

std_pre_l_df
```

```python
# 이상치 제거
for column in std_pre_l_df.columns:
    std_pre_l_df = std_pre_l_df[std_pre_l_df[column].between(-1.96, 1.96)]

std_pre_l_df
```

```python
# 이상치 제거 후 유지되는 인덱스 번호를 대조하여 타겟 데이터 가져오기
std_pre_l_df.loc[:, 'target'] = pre_l_df.loc[std_pre_l_df.index, 'Life expectancy ']
std_pre_l_df
```

```python
# 이상치 제거후 유지되는 인덱스를 대조하여 원본 데이터 이상치 제거
pre_l_df = pre_l_df.iloc[std_pre_l_df.index]
pre_l_df
```

```python
# 이상치 제거 후 인덱스 번호 초기화
pre_l_df.reset_index(drop=True, inplace=True)
```

```python
from statsmodels.api import OLS

# 데이터 세트 분리
# 피처, 타겟 데이터 분리
features, targets = pre_l_df.iloc[:, 1:], pre_l_df.iloc[:, 0]

# OLS 객체 생성
model = OLS(targets, features)
# 훈련 및 정보 출력
print(model.fit().summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
# VIF 확인
get_vif(features)
```

```python
from sklearn.decomposition import PCA

# 14번 반복하여 차원축소 객체 생성
for i in range(14):
    # 1차원~14차원까지 객체 생성
    pca = PCA(n_components=(i + 1))

    # 학습 및 테스트 데이터로 병합
    pca_train = pca.fit_transform(features)
    pca_test = pca.fit_transform(features)

    # 보존률 확인
    print(pca.explained_variance_ratio_.sum())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA

# 데이터 세트 분리
# 피처, 타겟 데이터 분리
features, targets = pre_l_df.iloc[:, 1:], pre_l_df.iloc[:, 0]

# 문제/정답 및 학습/테스트 데이터 분리
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# 선형 회귀 모델
l_r = LinearRegression()

# 파이프라인 구축
# 정규화, 차원축소, 선형회귀 모델
pipe = Pipeline([('std', StandardScaler()), ('pca', PCA(n_components=3)), ('l_r', l_r)])

# 훈련
pipe.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
# 예측
prediction = pipe.predict(X_test)
# 평가 (오차행렬)
get_evaluation(y_test, prediction)
```

```python
import pickle
import joblib

# 파이썬 객체를 직절화하고 파일에 저장
# joblib.dump(객체, '경로 및 이름')
# 객체를 불러와 사용할 수 있음
joblib.dump(pipe, './life.pkl')
```

```python
# 객체를 불러와 담기
life_model = joblib.load('./life.pkl')
# 모델로 예측
prediction = life_model.predict(X_test)
# 평가
get_evaluation(y_test, prediction)
```
